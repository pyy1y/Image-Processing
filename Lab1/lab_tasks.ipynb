### Task#2

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from PIL import Image
from skimage import data

img_path = "images/cameraman.tif"
pil_path = "images/lena_gray_256.tif"

if os.path.exists(img_path):
    img = cv2.imread(img_path)
else:
    cam = data.camera()
    img = cv2.cvtColor(cam, cv2.COLOR_GRAY2BGR)

plt.imshow(img)
plt.axis("off")
plt.show()

if os.path.exists(pil_path):
    img2 = Image.open(pil_path)
else:
    img2 = Image.fromarray(data.camera())

plt.imshow(img2, cmap=cm.Greys_r)
plt.axis("off")
plt.show()
```
