## Imports

```
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

## Task 2: Loading and Visualizing Images

### (a) Loading/reading image using OpenCV

```
img = cv2.imread('images/cameraman.tif')
plt.imshow(img)
plt.show()
```

### OpenCV: Convert BGR to RGB (optional for correct colors)

```
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

### (b) Loading/reading image using PIL

```
img2 = Image.open('images/lena_gray_256.tif')
plt.imshow(img2, cmap=cm.Greys_r)
plt.show()
```

## Task 3: Image Storing

### (a) Saving image to disk using OpenCV

```
cv2.imwrite('new_image.jpg', img)
```

### (b) Saving image to disk using PIL

```
img2.save('new_image2.jpg')
```

## Task 4: Display Image as an Array

### OpenCV image array

```
print(img.shape)
print(img)
```

### Convert PIL image to NumPy array

```
img_array = np.array(img2)
print(img_array.shape)
print(img_array)
```

## Assessment Practice: NumPy Operations on the Array

### Brightness adjustment

```
bright_img = img_array + 50
bright_img = np.clip(bright_img, 0, 255).astype(np.uint8)
plt.imshow(bright_img, cmap=cm.Greys_r)
plt.show()
```

### Negative image

```
negative_img = 255 - img_array.astype(np.uint8)
plt.imshow(negative_img, cmap=cm.Greys_r)
plt.show()
```

### Crop a region

```
cropped = img_array[50:150, 50:150]
plt.imshow(cropped, cmap=cm.Greys_r)
plt.show()
```

### Mean intensity

```
mean_intensity = np.mean(img_array)
mean_intensity
```
